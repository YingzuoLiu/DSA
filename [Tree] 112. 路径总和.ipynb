```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def hasPathSum(self, root: TreeNode, targetSum: int) -> bool:
        # 如果当前节点为空，返回 False
        if not root:
            return False
        
        # 如果当前节点是叶子节点，检查当前路径的总和是否等于 targetSum
        if not root.left and not root.right:
            return root.val == targetSum
        
        # 递归检查左子树和右子树
        return (self.hasPathSum(root.left, targetSum - root.val) or
                self.hasPathSum(root.right, targetSum - root.val))
```

树相关的调取函数：

1. 创建和初始化节点
TreeNode(val, left=None, right=None):
用于创建一个树的节点，val 是节点的值，left 和 right 分别是左子节点和右子节点。
示例：node = TreeNode(5) 创建了一个值为 5 的节点。
2. 遍历树
前序遍历 (Pre-order Traversal)

preorderTraversal(root): 访问当前节点，然后递归遍历左子树和右子树。
访问顺序：根节点 -> 左子树 -> 右子树
中序遍历 (In-order Traversal)

inorderTraversal(root): 递归遍历左子树，访问当前节点，然后递归遍历右子树。
访问顺序：左子树 -> 根节点 -> 右子树
后序遍历 (Post-order Traversal)

postorderTraversal(root): 递归遍历左子树和右子树，然后访问当前节点。
访问顺序：左子树 -> 右子树 -> 根节点
层序遍历 (Level-order Traversal)

levelOrder(root): 按层级顺序遍历树的节点，通常使用队列实现。
访问顺序：逐层从左到右访问节点。
3. 查找和判断
查找最近公共祖先 (Lowest Common Ancestor)

lowestCommonAncestor(root, p, q): 查找二叉树中两个节点 p 和 q 的最近公共祖先。
检查是否为二叉搜索树 (Binary Search Tree)

isValidBST(root): 判断一棵树是否为有效的二叉搜索树（BST）。
查找最大路径和

maxPathSum(root): 查找树中任意两个节点之间的最大路径和。
4. 修改树结构
翻转二叉树 (Invert Binary Tree)

invertTree(root): 翻转二叉树，使得树的左子树和右子树互换。
合并二叉树 (Merge Two Binary Trees)

mergeTrees(t1, t2): 合并两棵二叉树，将相同位置的节点值相加。
5. 特殊路径
计算路径和 (Path Sum)
hasPathSum(root, targetSum): 判断是否存在从根节点到叶子节点的路径，使得路径上的节点值之和等于给定的目标和。
二叉树的所有路径
binaryTreePaths(root): 列出所有从根节点到叶子节点的路径。
6. 构造树
根据前序和中序遍历构造二叉树

buildTree(preorder, inorder): 根据前序和中序遍历结果构造二叉树。
将有序数组转换为二叉搜索树

sortedArrayToBST(nums): 将升序数组转换为高度平衡的二叉搜索树。
7. 树的属性
树的最大深度 (Maximum Depth)

maxDepth(root): 计算二叉树的最大深度。
树的最小深度 (Minimum Depth)

minDepth(root): 计算二叉树的最小深度。
判断树是否平衡 (Balanced Binary Tree)

isBalanced(root): 判断二叉树是否是高度平衡的。
